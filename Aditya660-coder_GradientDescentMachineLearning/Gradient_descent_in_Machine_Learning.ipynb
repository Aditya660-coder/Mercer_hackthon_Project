```python
import numpy as np
```

```python
x=[2,3,5,6,7]
y =[26,45,25,36,56]
```

```python
def gradient_descent(x,y):
    m_curr=0
    b_curr=0
    learning_rate=0.003
    n = len(x)
    itr=10000
    for i in range(itr):
        y_pred= m_curr* x + b_curr
        md =-(2/n)*sum(x*(y-y_pred))
        bd =-(1/n)*sum(y-y_pred)
        m_curr= m_curr-(learning_rate*md)
        b_curr=b_curr-(learning_rate*bd)
        print("m_curr {}, b_curr {}, iteration {} ".format(m_curr,b_curr,i))
    
```

```python
print(x)
print(y)
x = np.array(x)
y = np.array(y)
print(x,y)
gradient_descent(x,y)
```
