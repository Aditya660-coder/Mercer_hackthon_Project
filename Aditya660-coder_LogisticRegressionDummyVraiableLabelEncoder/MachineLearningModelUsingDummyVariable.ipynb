```python
import pandas as pd
df = pd.read_csv("housepricing.csv")
```

```python
df
```

# Using dummy Variables we can change the town values to the classified numeriacal values .

```python
dummy = pd.get_dummies(df.town)
```

```python
dummy
```

# If we don't want dummy variable trap.We have to remove one dummy column

```python
final = dummy.drop(['west windsor'],axis='columns')
final
```

```python
merged = pd.concat([final,df],axis='columns')
```

```python
merged
```

```python
new_df = merged.drop(['town'],axis='columns')
```

```python
new_df
```

```python
x=new_df[['monroe township','robinsville','area']]
```

```python
x
```

```python
y = new_df.price
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict([[1,0,2500]])[0]
```

```python
model.score(x,y)
```

```python
import pickle
with open('model.pickle' ,'wb') as f:
    pickle.dump(model,f)
```

```python
with open('model.pickle','rb') as f:
    md = pickle.load(f)
```
